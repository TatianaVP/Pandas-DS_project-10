**3. Произведите агрегации с маркой и моделью:**

Отобразите медианные штрафы в разрезе только марок (марка – медианный штраф внутри этой марки);

```python
df.groupby("Make").agg({'Fines':'median'})
```

Отобразите медианные штрафы, сгруппированные по маркам и моделям (комбинация марка и модель – медианный штраф этой комбинации);

```python
df.groupby(["Make", "Model"]).agg({'Fines':'median'})
```

Отобразите количество штрафов, имеющихся в группировках по маркам и моделям, чтобы понять, можно ли доверять полученным медианным значениям;

```python
df.groupby(["Make", "Model"]).agg({'Fines':'count'})
```

Отобразите минимальные и максимальные штрафы, сгруппированные по маркам и моделям для того, чтобы лучше понять вариацию;

```python
df.groupby(["Make", "Model"]).agg({'Fines':'min'})
```

```python
df.groupby(["Make", "Model"]).agg({'Fines':'max'})
```

Отобразите стандартное отклонение штрафов, сгруппированных по маркам и моделям для того, чтобы лучше понять вариацию.

```python
df.groupby(["Make", "Model"]).agg({'Fines':'std'})
```

**4.Произведите агрегации с регистрационным номером автомобиля:**

Отобразите регистрационные номера автомобилей, сгруппированные по количеству штрафов в порядке убывания (мы хотим найти тех, кто чаще всего нарушал правила дорожного движения);

```python

fine_counts = df.reset_index().groupby('CarNumber', as_index=False).agg({'Fines':'count'}).sort_values('Fines', ascending=False)
fine_counts
```

Выберите из исходного датафрейма все строки, соответствующие регистрационному номеру с самым большим количеством нарушений (мы хотим проанализировать эти данные более подробно);

```python
df.loc[fine_counts.iloc[0,0]]
```

Отобразите регистрационные номера автомобилей, сгруппированные по сумме штрафов в порядке убывания (мы хотим найти тех, кто заплатил самую большую сумму штрафов);

```python
fines = df.reset_index().groupby('CarNumber', as_index=False).agg({'Fines':'sum'}).sort_values('Fines', ascending=False)
fines
```

Выберите из исходного датафрейма все строки, соответствующие регистрационному номеру с самой большой суммой штрафов (мы хотим проанализировать эти данные более подробно);

```python
df.loc[fines.iloc[0,0]]
```

Вывести на экран таблицу, отвечающую на вопрос: есть ли регистрационные номера, которые были установлены на разных моделях автомобилей?

```python
number = df.reset_index().groupby('CarNumber', as_index=False).agg({'Model':'count'}).value_counts('Model').reset_index().sort_values('Model')
number.columns = ['Number_Models', 'Number_Cars']
number
```
