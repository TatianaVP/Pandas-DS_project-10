Ваши задачи в рамках этого упражнения в деталях.

Используя функцию read_csv:
1.   Отфильтруйте строки с индексами 2 и 3 с помощью аргумента skiprows (мы заведомо знаем, что эти наблюдения были фиктивными);
Отфильтруйте последние 2 строки таблицы с помощью аргумента skipfooter (мы знаем, что эти наблюдения тоже были фиктивными);
Присвойте столбцам следующие имена: datetime, user;
Используйте datetime в качестве столбца индекса.
2.   Переименуйте datetime в date_time.
3.   Используя метод to_csv:
Используйте ; в качестве разделителя;
Сохраните данные в файл с именем feed-views-semicolon.log.

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Пункт 1
df = pd.read_csv(
    '/content/drive/MyDrive/Школа 21/DS_project10-0/Task 1/feed-views.log',
    sep='\t',
    engine='python',
    header=None,
    skiprows=[2, 3], #пропустить 2 и 3 строки
    skipfooter=2, #количество последних строк, которые нужно пропустить
    names=['datetime', 'user'],
    index_col='datetime'
    ) 
df
```

```python
# Пункт 2
df.index.names = ['date_time']
df
```

```python
# Пункт 3
df.to_csv('/content/drive/MyDrive/Школа 21/DS_project10-0/Task 1/feed-views-semicolon.log', sep=';') #index=False
```
